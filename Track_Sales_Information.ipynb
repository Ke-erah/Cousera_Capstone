# Selecting Albums to Purchase

A query to find out which genres sell the most tracks in the USA

```python
%%sql
WITH usa AS (SELECT c.customer_id, c.country, i.invoice_id, i.total
FROM customer AS c INNER JOIN invoice AS i
ON c.customer_id=i.customer_id
WHERE country='USA'), 

track_qty_usa AS (SELECT u.*, il.track_id, il.quantity
FROM usa AS u
LEFT JOIN invoice_line AS il
ON u.invoice_id=il.invoice_id),

genre_id_usa AS (SELECT tqu.*, t.genre_id
FROM track_qty_usa AS tqu 
LEFT JOIN track AS t
ON tqu.track_id=t.track_id),

genre_usa AS (SELECT giu.country, SUM(giu.quantity) AS quantity, g.name AS genre
FROM genre_id_usa AS giu
INNER JOIN genre AS g
ON giu.genre_id=g.genre_id
GROUP BY g.name)

SELECT genre, quantity, 
CAST(quantity AS Float)/(SELECT SUM(quantity) FROM genre_usa) AS percentage
FROM genre_usa AS gu
ORDER BY quantity DESC
```

From the above table/finding, it can be seen that the order of popularity of the genres in the USA (based on the new tracks) is as follows:
1. Punk
2. Blues
3. Pop
4. Hip-hop

Hence, the recommendation would be to select artists: 
1. Red Tone(Punk)
2.Slim Jim Bites(Blues)
3.Meteor and the Girls(Pop)

# Analyzing Employees Sales Performance

```python
%%sql
SELECT e.first_name||' '||e.last_name employee_name,
e.title employee_title, e.hire_date, sum(i.total) total_sales
FROM employee AS e
INNER JOIN customer AS c
ON e.employee_id=c.support_rep_id
INNER JOIN invoice AS i
ON c.customer_id=i.customer_id
WHERE e.title='Sales Support Agent'
GROUP BY employee_name
```

# Analyzing Sales by Country

```python
%%sql
CREATE VIEW sales_data AS
SELECT c.customer_id, c.country, i.total, il.quantity FROM customer c
INNER JOIN invoice i
ON c.customer_id=i.customer_id
INNER JOIN invoice_line il
ON i.invoice_id=il.invoice_id
```

```python
%%sql
DROP VIEW sales_data
```

```python
%%sql
CREATE VIEW sales_data AS 
SELECT COUNT(DISTINCT(c.customer_id)) customer_count,
COUNT(DISTINCT(i.invoice_id)) invoice_count,
c.country, SUM(il.unit_price) total_sales FROM customer c
INNER JOIN invoice i
ON c.customer_id=i.customer_id
INNER JOIN invoice_line il
ON i.invoice_id=il.invoice_id
GROUP BY country
```

```python
%%sql
CREATE VIEW summary AS 
SELECT new_country, SUM(customer_count) customer_count, 
SUM(total_sales) total_sales, total_sales/customer_count avg_sales,
total_sales/invoice_count ag_order FROM 
(
    SELECT sd.*, 
    CASE
    WHEN sd.customer_count=1 THEN 'Other'
    Else sd.country
    END AS new_country
    FROM sales_data sd)
GROUP BY new_country
ORDER BY total_sales DESC
```

```python
%%sql
SELECT new_country AS country_name, customer_count, total_sales,
avg_sales, ag_order AS avg_order FROM 
(
    SELECT s.*, 
    CASE
    WHEN s.new_country='Other' THEN 1
    ELSE 0
    END AS sort
    FROM summary s
)
ORDER BY sort 
```

# Albums vs Individual Tracks

```python
%%sql
CREATE VIEW track_sales AS 
SELECT * FROM invoice_line il INNER JOIN track t ON il.track_id=t.track_id
```

```python
%%sql
SELECT * FROM track_sales LIMIT 1;
```

```python
%%sql
SELECT track_id FROM track_sales
EXCEPT
SELECT album_id FROM track_sales
```

```python
%%sql

WITH invoice_first_track AS
    (
     SELECT
         il.invoice_id invoice_id,
         MIN(il.track_id) first_track_id
     FROM invoice_line il
     GROUP BY 1
    )

SELECT
    album_purchase,
    COUNT(invoice_id) number_of_invoices,
    CAST(count(invoice_id) AS FLOAT) / (
                                         SELECT COUNT(*) FROM invoice
                                      ) percent
FROM
    (
    SELECT
        ifs.*,
        CASE
            WHEN
                 (
                  SELECT t.track_id FROM track t
                  WHERE t.album_id = (
                                      SELECT t2.album_id FROM track t2
                                      WHERE t2.track_id = ifs.first_track_id
                                     ) 

                  EXCEPT 

                  SELECT il2.track_id FROM invoice_line il2
                  WHERE il2.invoice_id = ifs.invoice_id
                 ) IS NULL
             AND
                 (
                  SELECT il2.track_id FROM invoice_line il2
                  WHERE il2.invoice_id = ifs.invoice_id

                  EXCEPT 

                  SELECT t.track_id FROM track t
                  WHERE t.album_id = (
                                      SELECT t2.album_id FROM track t2
                                      WHERE t2.track_id = ifs.first_track_id
                                     ) 
                 ) IS NULL
             THEN "yes"
             ELSE "no"
         END AS "album_purchase"
     FROM invoice_first_track ifs
    )
GROUP BY album_purchase;
```
